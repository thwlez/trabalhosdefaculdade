```python
from ast import Continue

print('======================= Bem vindos a plataforma de Serviços de Limpeza do Thales =======================')

#Início da função metragem_limpeza()
def metragem_limpeza():
  print('======================= Menu 1 de 3 - Metragem de Limpeza =======================\n' + 'Insira a metragem da casa (valor em m²):')
  while True:
    try:
      entrada_metragem = int(input('>> '))
      if (30 <= entrada_metragem) and (entrada_metragem < 300) :
        return entrada_metragem + 60 + (entrada_metragem * 0.3)
      elif 300 <= entrada_metragem < 700:
        return entrada_metragem + 120 + (entrada_metragem * 0.5)
      else:
        print('!! Não aceitamos casas com metragem menor que 30m² ou maiores que 700m² !!')
    except ValueError:
      print('!! Digite apenas valores inteiros !!')
#Fim da função 

#Início da função tipo_limpeza()
def tipo_limpeza():
  print('============================== Menu 2 de 3 - Tipo de Limpeza ==============================\n' + 'Insira o tipo de limpeza desejado:\n' + 
        'B – Básica - Indicada para sujeiras semanais ou quinzenais\n' + 
        'C – Completa - Indicada para sujeiras antigas e/ou não rotineiras')
  while True:
    entrada_tipo = input('>> ')
    entrada_tipo = entrada_tipo.upper()
    entrada_tipo = entrada_tipo.strip()
    if entrada_tipo == 'B':
      return 1.00
    elif entrada_tipo == 'C':
      return 1.30
    else:
      print('!! Insira somente o código dos tipos disponíveis :  B  ou  C !!')
#Fim da função

#Início da adicional_limpeza()
def adicional_limpeza():
  print('============================= Menu 3 de 3 - Adicional de Limpeza =============================')
  print('0 -  Não desejo mais nada (encerrar) ; \n' + 
        '1 -  Passar dez (10) peças de roupas -  R$ 10.00 ;\n' + 
        '2 -  Limpeza de 1 Forno/Micro-ondas  -  R$ 12,00 ;\n' +
        '3 -  Limpeza de 1 Geladeira/Freezer  -  R$ 20,00 .')

  acumulador_limpeza = 0
  while True:
    entrada_adicional = input('Deseja algo a mais?\n' + '>> ')
    if entrada_adicional == '0':
      return acumulador_limpeza
    elif entrada_adicional == '1':
      acumulador_limpeza += 10.00
      continue
    elif entrada_adicional == '2':
      acumulador_limpeza += 12.00
      continue
    elif entrada_adicional == '3':
      acumulador_limpeza += 20.00
      continue
    else:
      print ('!! Código inválido, digite somente algum dos códigos da tabela (0, 1, 2, 3) !!')
#Fim da função 

#Chamando as funções e fazendo o somatório final
metragem = metragem_limpeza()
tipo = tipo_limpeza()
adicionais = adicional_limpeza()

total = metragem * tipo + adicionais 

print('\n*************************************************************************************\n')
print('O total é de R$ {:.2f} (metragem: {:.2f} * tipo: {:.2f} + adicional(is): {:.2f})'.format(total,metragem,tipo,adicionais))
```
