```python
print('----------  Bem-vindo ao APP de vendas do Thales!  ----------')
valor_produto = float(input('Digite o valor do produto desejado: \n' + '>>')) #valor do produto comprado
qtd_embalagem = int(input('Digite a quantidade de embalagens desejada: \n' + '>>')) #quantidade de embalagens
valor_frete = 0 #acumulador do frete

if (qtd_embalagem >= 0) and (qtd_embalagem < 11):
  valor_frete += 30
elif (qtd_embalagem >= 11) and (qtd_embalagem < 101):
  valor_frete += 60
elif (qtd_embalagem >= 101) and (qtd_embalagem < 1001):
  valor_frete += 120
else:
  valor_frete += 240

valor_parcial = valor_produto * qtd_embalagem
valor_final = valor_parcial + valor_frete

print('Valor total s/ o frete: R$ {:.2f}'.format(valor_parcial))
print('Valor total c/ o frete: R$ {:.2f}\n'.format(valor_final)+'(frete de R$ {:.2f})'.format(valor_frete))
```
